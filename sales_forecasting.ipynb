```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
#pie
df['IsHoliday_x'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Holiday Percentage")
plt.ylabel('') 
plt.show()
```

```python
#bar
df['Type'].value_counts().plot(kind='bar')
plt.title("Count of Store Types")      
plt.show()
```

```python
#hist
df['MarkDown1'].plot(kind='hist')
plt.title("Distribution of MarkDown1") # Recommended for clarity
plt.show()
```

```python
#box
df.boxplot(column='MarkDown1')
plt.title("Boxplot of MarkDown1")
plt.show()
```

```python
#line
df.groupby('Type')['MarkDown1'].mean().plot(kind='line')
plt.title("MarkDown1 Trend Over Time")
plt.xlabel("Type")
plt.ylabel("Average MarkDown1")
plt.show()
```

```python
#scatter plot
df.plot(kind='scatter', x='MarkDown1', y='MarkDown2')
plt.title("Scatter Plot: MarkDown1 vs MarkDown2")
plt.xlabel("MarkDown1")
plt.ylabel("MarkDown2")
plt.show()
```

```python
#Heatmap
sn.heatmap(df[['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.show()
```

```python
sn.countplot(x='Type', data=df)
plt.title("Number of Rows per Store Type")
plt.xlabel("Store Type")
plt.ylabel("Count")
plt.show()
```

```python
#pair plot
columns = df[['MarkDown1', 'MarkDown2', 'MarkDown3']]
sn.pairplot(columns)
plt.show()
```

```python
#violin plot
sn.violinplot(x='Type', y='MarkDown1', data=df.dropna(subset=['MarkDown1']))

plt.title("Distribution Shape of MarkDown1 by Store Type")
plt.xlabel("Store Type")
plt.ylabel("MarkDown1 Value")
plt.show()
```

```python
#Stacked Area Chart
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
weekly_markdowns = df.groupby('Date')[markdown_cols].mean()
weekly_markdowns.plot(kind='area', figsize=(12, 6), alpha=0.7)

plt.title("Cumulative Contribution of MarkDowns Over Time")
plt.xlabel("Date")
plt.ylabel("Total Combined Average")
plt.legend(title="Markdown Type", loc='upper left')
plt.show()
```

```python
df.groupby('Type')[['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']].mean().plot(kind='bar')
plt.title('markdown by type')
plt.ylabel('Markdown')
plt.show()
```

```python
df.groupby('IsHoliday_x')[['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']].mean().plot(kind='bar')
plt.title('markdown by Holiday')
plt.ylabel('Markdown')
plt.show()
```
